```python
import numpy as np
from numpy import asarray
from skimage import io
from skimage import color
import matplotlib.pyplot as plt
from PIL import Image


image=io.imread(input("Enter image name: \n"))

print("Image Type",type(image),'\t Image Shape: ', image.shape)

print('''Enter the Option:
            1. Load Image
            2. Edit Image
            3.  Quit Image
        ''')

flag1=True
while(flag1):
    option1=int(input("Choice (1/2/3): "))
    if option1==1:
        print("Show Image\n")
        plt.imshow(image)
        plt.show()
    elif option1==2:
        print("Edit Image\n")
        flag2=True
        print('''Enter the Option:
                    1. Horizontal Flip
                    2. Vertical Flip
                    3. Invert Colors
                    4. Convert to Grayscale
                    5. Quit
                ''')
        while(flag2):
            option2=int(input("Choice (1/2/3/4/5): "))
            if option2==1:
                print("Horizontal Flip\n")           #Horizontal Flip
                np_image_flipped=image[:,::-1]       
                plt.imshow(np_image_flipped)
                plt.show()
                print('''Enter the Option:
                    1. Horizontal Flip
                    2. Vertical Flip
                    3. Invert Colors
                    4. Convert to Grayscale
                    5. Quit
                ''')
                flag2=True
            elif option2==2:
                print("Vertical Flip\n")             #Vertical Flip
                copy1=Image.fromarray(np.flipud(image))
                plt.imshow(copy1)
                plt.show()
                print('''Enter the Option:
                    1. Horizontal Flip
                    2. Vertical Flip
                    3. Invert Colors
                    4. Convert to Grayscale
                    5. Quit
                ''')
                flag2=True
            elif option2==3:
                np_image=np.invert(image)                     #Invert
                plt.imshow(np_image)
                plt.show()
                print('''Enter the Option:
                    1. Horizontal Flip
                    2. Vertical Flip
                    3. Invert Colors
                    4. Convert to Grayscale
                    5. Quit
                ''')
                flag2=True
            elif option2==4:
                print("Convert to Grayscale\n")       #Grayscale and Balck&White
                imgGray=color.rgb2gray(image)
                io.imshow(imgGray)
                plt.show()
                print('''Enter the Option:
                    1. Horizontal Flip
                    2. Vertical Flip
                    3. Invert Colors
                    4. Convert to Grayscale
                    5. Quit
                ''')
                flag2=True
            elif option2==5:
                print("Quit Editing\n")
                break
            else:
                print("Wrong Choice! Choice (1/2/3/4/5): ")
    elif option1==3:
        print("Quit\n")
        break
    else:
        print("Wrong Choice! Choice (1/2/3): ")
```
